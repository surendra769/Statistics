```python
contingency_tab = pd.crosstab(df1['What is your most preferred way of investing ?'],
                             df1['Which of these describes your personal income per annum (in lakh) ?'])
```

```python
contingency_tab
```

```python
sns.heatmap(contingency_tab, cmap = 'Greens', annot = True);
```

```python
from scipy.stats import chi2_contingency
stat, p, dof,expected = chi2_contingency(contingency_tab)
```
